```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import cv2 
```

```python
img = cv2.imread("BW.png")
np.shape(img)
```

```python
plt.imshow(img)
plt.show()
```

```python
gray_img = cv2.cvtColor(img , cv2.COLOR_BGR2GRAY)
```

```python
shape = np.shape(gray_img)
```

```python

# idmi bu ksmda direk olarak rowlara eriim salayabiliyoruz.
counter = 0
for i in gray_img:
    counter += 1
    print(i)
    if counter == 10:
        break;
    
```

```python
len(gray_img[0])
```

```python
collected_point = []
counter = 0


for i in range(shape[0]):
    for j in range(shape[1]):
        
        value = gray_img[i,j]
        
        if value > 100:
            collected_point.append([i,j])
            
collected_point
```

```python
def disstance_fun(x,y):
    return np.sum(np.square(np.subtract(x, y))**0.5)

filtered_collected_items = []
counter = 0

for i in collected_point:
    
    if len(filtered_collected_items) == 0: 
        filtered_collected_items.append(i)
        continue
        else:
            disstance_fun(i,collected_point[counter])
            counter += 1
    #print(collected_point[i])
    
    
```
